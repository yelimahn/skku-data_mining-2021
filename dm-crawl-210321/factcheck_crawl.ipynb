```python
# Text 가져오는 함수
def ele_to_text(url):
    driver.get(url)
    
    category = driver.find_element_by_xpath('//*[@id="content_detail"]/div/div[3]/div[1]/div/div[2]/div[2]/ul/li[2]').text
    speaker = driver.find_element_by_xpath('//*[@id="content_detail"]/div/div[3]/div[1]/div/div[1]/p').text
    title = driver.find_element_by_xpath('//*[@id="content_detail"]/div/div[3]/div[1]/div/div[2]/div[1]/p[1]/a').text
    source = driver.find_element_by_xpath('//*[@id="content_detail"]/div/div[3]/div[1]/div/div[2]/div[2]/ul/li[1]').text
    veracity = driver.find_element_by_xpath('//*[contains(@id,"meter-item-")]/div[2]').text
    logo_url = driver.find_element_by_xpath('//*[contains(@id,"score-")]/div/div[1]/div[2]/ul/li/img').get_attribute('src')
    
    return [category, speaker, title, source, veracity, logo_url]
```

```python
# url text 가져오기
from tqdm import tqdm
import time

brief_data = [['category', 'speaker', 'title', 'source', 'veracity', 'logo_url']]
for url in tqdm(brief_url):
    if url is not None: 
        data = ele_to_text(url)
        brief_data.append(data)
        time.sleep(3)
```

```python
import csv

def make_csv(crawl_list):
  #file 변수에서 csv파일 오픈
  file = open("factcheck.csv", 'w', newline='')
  csvfile = csv.writer(file)
  #줄 별로 읽어서 csv파일에 저장
  for row in crawl_list :
    csvfile.writerow(row)
  #파일 닫기
  file.close()
```

```python
make_csv(brief_data)
print("csv file uploaded!!")
```
