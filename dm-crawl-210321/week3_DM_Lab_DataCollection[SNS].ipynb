## Youtube

- Youtube API document : https://developers.google.com/youtube/v3/docs/search/list?hl=ko
- Google Platform : https://console.cloud.google.com/?hl=ko

```python
#youtube api
from googleapiclient.discovery import build
```

```python
def load_api(token):
    DEVELOPER_KEY = token
    YOUTUBE_API_SERVICE_NAME = "youtube"
    YOUTUBE_API_VERSION = "v3"
    return build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,developerKey=DEVELOPER_KEY)
```

```python
tokens = "Replace Me!"
```

```python
youtube = load_api(tokens)
```

## 검색어 기반 유튜브 정보 검색



```python
search_response = youtube.search().list(  
      q= "BTS", 
      publishedAfter = '2021-03-01T00:00:00Z', #RFC 3339
      publishedBefore = '2021-03-02T00:00:00Z',
      part= 'id,snippet',
      type = 'video',
      order = 'relevance',
      maxResults = 10,
      regionCode = 'US'
      ).execute()
```

```python
search_response
```

```python
len(search_response['items'])
```

```python
search_response['items'][0]
```

```python
print(search_response['items'][0]['id']['videoId']) #videoid
print(search_response['items'][0]['snippet']['title']) #video title
print(search_response['items'][0]['snippet']['description']) #video description
print(search_response['items'][0]['snippet']['publishedAt']) #upload_date
```

```python
search_response['nextPageToken']
```

```python
search_response_next = youtube.search().list(  
      q= "BTS", 
      publishedAfter = '2021-03-01T00:00:00Z', #RFC 3339
      publishedBefore = '2021-03-02T00:00:00Z',
      part= 'id,snippet',
      type = 'video',
      order = 'relevance',
      maxResults = 10,
      regionCode = 'US',
      pageToken = search_response['nextPageToken']
      ).execute()
```

## 유튜브 영상에 달린 댓글 모음 

```python
vidio_id = search_response['items'][0]['id']['videoId']
vidio_id
```

```python
cmt_response = youtube.commentThreads().list(
    part= 'snippet',
    videoId = vidio_id,
    maxResults=10
    ).execute()
```

```python
len(cmt_response['items'])
```

```python
cmt_response['items'][0]
```

```python
print(cmt_response['items'][3]['snippet']['topLevelComment']['snippet']['textDisplay'])
```
