```
# Text 가져오는 함수
def ele_to_text(url):
    driver.get(url)
    
    title = (driver.find_element_by_xpath('//*[@id="articleTitle"]')).text
    date = (driver.find_element_by_xpath('//*[@id="main_content"]/div[1]/div[3]/div/span[1]')).text
    publication = (driver.find_element_by_xpath('//*[@id="footer"]/address[1]/a')).text
    body = (driver.find_element_by_xpath('//*[@id="articleBodyContents"]')).text
    
    return title, date, publication, body
```

```
# url text 가져오기
from tqdm import tqdm
import time

brief_data = [['title','date','publication',body]]
for url in tqdm(brief_url):
    if url is not None: 
        #driver.get(url)
        data = ele_to_text(url)
        brief_data.append(data)
        time.sleep(3)
```

```
import csv
import time

def make_csv(crawl_list):
  #file 변수에서 csv파일 오픈
  file = open("naver_news_상장_crawl", 'w', newline='')
  csvfile = csv.writer(file)
  #줄 별로 읽어서 csv파일에 저장
  for row in crawl_list :
    csvfile.writerow(row)
  #파일 닫기
  file.close()
```

```
make_csv(brief_data)
print("csv file uploaded!!")
```
