Created a directed graph

*  nodes : 20대 국회의원
*  edges : 공동으로 입법 발의안을 냄




```python
# PUBL_PROPOSER에 공동으로 여러 사람의 이름이 들어가므로, split해주어 리스트 형식으로 다시 집어넣음.
for i in range(df_suggest.shape[0]):
  df_suggest['PUBL_PROPOSER'][i] = df_suggest.PUBL_PROPOSER.iloc[i].split(',')
```

```python
df_suggest.head()
```

```python
# graph 만들기
g = nx.DiGraph()
```

```python
# people_20.csv에 나온 의원들을 node로 설정
people = []

for i in range(df_ppl.shape[0]):
  people.append(df_ppl['의원명'][i])
```

```python
g.add_nodes_from(people)
```

```python
g.nodes
```

```python
# 리스트 형식으로 변수에 집어 넣음.

rst_proposer = list(df_suggest['RST_PROPOSER'])
publ_proposer = list(df_suggest['PUBL_PROPOSER'])
bill_index = list(df_suggest['BILL_index'])
```

```python
# 공동으로 입법 발의안을 만든 사람들의 edge를 만들어 g에 설정

for i in range(df_suggest.shape[0]):
  for j in range(len(publ_proposer[i])):
    g.add_edge(rst_proposer[i], publ_proposer[i][j])
    g.edges[rst_proposer[i], publ_proposer[i][j]]['BILL_index'] = bill_index[i]
```

```python
# 정상적으로 edge가 생성됨.
g.edges
```

```python
# 20대 국회의원들
len(g.nodes)
```

```python
# 20대 국회의원들이 입법 발의안으로써 교류한 edge의 개수
len(g.edges)
```

Analyze a graph

```python
def viewer(_dict_data, col_name):
    result = pd.DataFrame().from_dict(_dict_data, orient='index', columns=[col_name])
    return df_ppl.set_index('의원명').join(result, how='right').sort_values(col_name, ascending=False).head(10)
```

Page Rank

```python
pagerank = nx.pagerank(g)
```

```python
pagerank
```

```python
pagerank = viewer(pagerank, 'pagerank')
pagerank
```

Node Centrality

```python
degree_centrality = nx.degree_centrality(g)
degree_centrality = viewer(degree_centrality, 'degree_centrality')
degree_centrality
```

```python
closeness_centrality = nx.closeness_centrality(g)
closeness_centrality = viewer(closeness_centrality, 'closeness_centrality')
closeness_centrality
```

```python
betweenness_centrality = nx.betweenness_centrality(g,  normalized=True)
betweenness_centrality = viewer(betweenness_centrality, 'betweenness_centrality')
betweenness_centrality
```

```python
eigenvector_centrality = nx.eigenvector_centrality(g)
eigenvector_centrality = viewer(eigenvector_centrality, 'eigenvector_centrality')
eigenvector_centrality
```

Community

```python
from networkx.algorithms import community
comp = community.girvan_newman(g)
```

```python
communities = tuple(sorted(c) for c in next(comp))
```

```python
for i, comm_nodes in enumerate(communities):
    print('community {}th'.format(i))
    print(comm_nodes, end='\n')
```
