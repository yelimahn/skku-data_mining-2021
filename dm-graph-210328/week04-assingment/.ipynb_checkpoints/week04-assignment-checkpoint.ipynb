```python
import networkx as nx
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
from tqdm.notebook import tqdm
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt


%matplotlib inline
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 100)
```

```python
# data load

df_suggest = pd.read_csv('./suggest_20.csv', encoding='cp949')
df_ppl = pd.read_csv('./people_20.csv', encoding='cp949')
```

```python
df_suggest.columns
```

# 1. task1

```python
# node 연결

node_list = []
edge_list = []
for i in tqdm(range(len(df_suggest))):
    rst = df_suggest.RST_PROPOSER.iloc[i]
    pul_list = df_suggest.PUBL_PROPOSER.iloc[i].split(',')
    committee = df_suggest.COMMITTEE_index.iloc[i]
    for pul in pul_list:
        node_list.append((rst,pul))
        edge_list.append(committee)
```

```python
df = pd.DataFrame()
df['node1'] = [x for x,y in node_list]
df['node2'] = [y for x,y in node_list]
df['edge'] = [x for x in edge_list]
```

```python
df
```

```python
df_ppl['party_index'] = df_ppl.party_index.apply(lambda x : str(x))
```

```python
g = nx.DiGraph(df.loc[:, ('node1', 'node2')].dropna().values.tolist()) # [(vs_id, vs_id)]
```

```python
g.edges()
```

```python
nx.set_node_attributes(g, dict(df_ppl.set_index('의원명')['party_index']), 'party')
```

```python
nx.set_edge_attributes(g, df.set_index(['node1', 'node2'])['edge'], 'committee') # ex. {(1, 3): val}
```
